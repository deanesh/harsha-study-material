| 11        | **Sets**                         | Union                              | [ $A \cup B = \{ x : x \in A \text{ or } x \in B \}$ ]                                                                    | $A, B$ are sets; $x$ is an element belonging to $A$ or $B$        |
|           |                                  | Intersection                       | [ $A \cap B = \{ x : x \in A \text{ and } x \in B\}$ ]                                                                   | $A, B$ are sets; $x$ is an element common to both $A$ and $B$     |
|           |                                  | Complement                         | [ $A' = U - A$ ]                                                                                                          | $A$ is a subset of universal set $U$; $A'$ is complement of $A$  |
|           |                                  | Cardinality                        | [ $n(A \cup B) = n(A) + n(B) - n(A \cap B)$ ]                                                                             | $n(X)$ is number of elements in set $X$                           |
| 11        | **Relations & Functions**         | Function                            | A relation where every input has exactly one output                                                                         | Input = domain element, Output = range element                     |
|           |                                  | Domain and Range                   | Set of inputs (domain), outputs (range)                                                                                   | Domain = all possible inputs; Range = all possible outputs        |
|           |                                  | Types of Functions                 | One‑one, onto, bijective                                                                                                  | Describes function properties                                      |
| 11        | **Trigonometry**                 | Identities                          | [ $\sin^2 x + \cos^2 x = 1$, $1 + \tan^2 x = \sec^2 x$, $1 + \cot^2 x = \csc^2 x$ ]                                    | $x$ is the angle in radians or degrees                             |
|           |                                  | Angle Formulas                     | [ $\sin(A \pm B) = \sin A \cos B \pm \cos A \sin B$ ]                                                                    | $A, B$ are angles                                                  |
|           |                                  | Double Angle                       | [ $\sin 2A = 2 \sin A \cos A$, $\cos 2A = \cos^2 A - \sin^2 A$ ]                                                         | $A$ is angle                                                       |
| 11        | **Complex Numbers**              | General Form                       | [ $z = a + ib$ ]                                                                                                          | $a, b$ are real numbers; $i = \sqrt{-1}$                           |
|           |                                  | Conjugate                          | [ $\bar{z} = a - ib$ ]                                                                                                    | Conjugate of $z$                                                   |
|           |                                  | Modulus                            | [ $|z| = \sqrt{a^2 + b^2}$ ]                                                                                              | Magnitude of complex number $z$                                   |
| 11        | **Quadratic Equations**          | Roots                              | [ $x = \frac{-b \pm \sqrt{b^2 - 4ac}}{2a}$ ]                                                                               | $a, b, c$ are coefficients of $ax^2 + bx + c = 0$                  |
|           |                                  | Discriminant                       | [ $D = b^2 - 4ac$ ]                                                                                                       | $D$ determines nature of roots                                     |
|           |                                  | Sum & Product of Roots            | [ $\alpha + \beta = -\frac{b}{a}$, $\alpha\,\beta = \frac{c}{a}$ ]                                                      | $\alpha, \beta$ are roots                                         |
| 11        | **Sequence & Series**            | AP nth Term                        | [ $a_n = a + (n - 1)d$ ]                                                                                                  | $a$ = first term, $d$ = common difference, $n$ = term number      |
|           |                                  | AP Sum                             | [ $S_n = \frac{n}{2} [2a + (n - 1)d]$ ]                                                                                   | Sum of first $n$ terms in AP                                       |
|           |                                  | GP nth Term                        | [ $a_n = ar^{\,n-1}$ ]                                                                                                     | $a$ = first term, $r$ = common ratio                              |
|           |                                  | GP Sum (finite)                    | [ $S_n = a \frac{r^n - 1}{r - 1}$, $r \ne 1$ ]                                                                            | Sum of first $n$ terms in GP                                       |
|           |                                  | GP Sum (infinite)                 | [ $S_{\infty} = \frac{a}{1 - r}$, $|r| < 1$ ]                                                                              | Sum of infinite GP                                                 |
| 11        | **Permutation & Combination**    | Permutation                         | [ $_nP_r = \frac{n!}{(n - r)!}$ ]                                                                                         | Number of arrangements of $r$ objects from $n$                     |
|           |                                  | Combination                        | [ $_nC_r = \frac{n!}{r!\,(n - r)!}$ ]                                                                                     | Number of ways to choose $r$ objects from $n$                      |
|           |                                  | Property                           | [ $_nC_r = _nC_{n - r}$ ]                                                                                                  | Symmetry property                                                   |
| 11        | **Binomial Theorem**            | Expansion                           | [ $(x + a)^n = \sum_{k=0}^{n} \binom{n}{k} x^{\,n-k} a^k$ ]                                                                | $n$ = power; $k$ = index; $\binom{n}{k}$ = binomial coefficient    |
|           |                                  | General Term                       | [ $T_{k+1} = \binom{n}{k} x^{\,n-k} a^k$ ]                                                                                 | Term number $k+1$                                                   |
| 11        | **Straight Lines**             | Slope                              | [ $m = \frac{y_2 - y_1}{x_2 - x_1}$ ]                                                                                      | Slope of line through points $(x_1,y_1)$ and $(x_2,y_2)$          |
|           |                                  | Point‑Slope Form                  | [ $y - y_1 = m(x - x_1)$ ]                                                                                                 | Equation of line with slope $m$ through $(x_1,y_1)$                |
|           |                                  | Slope‑Intercept Form              | [ $y = mx + c$ ]                                                                                                          | $m$ = slope, $c$ = y‑intercept                                     |
|           |                                  | Distance                           | [ $\sqrt{(x_2 - x_1)^2 + (y_2 - y_1)^2}$ ]                                                                                 | Distance between two points                                        |
|           |                                  | Angle Between Lines               | [ $\tan \theta = \left| \frac{m_1 - m_2}{1 + m_1 m_2} \right|$ ]                                                         | $\theta$ = angle between lines with slopes $m_1, m_2$              |
| 11        | **Conic Sections**             | Circle                             | [ $(x - h)^2 + (y - k)^2 = r^2$ ]                                                                                          | Center at $(h,k)$; radius = $r$                                     |
|           |                                  | Parabola                           | [ $y^2 = 4ax$, $x^2 = 4ay$ ]                                                                                               | $a$ = distance from vertex to focus                                 |
|           |                                  | Ellipse                            | [ $\frac{x^2}{a^2} + \frac{y^2}{b^2} = 1$ ]                                                                                | $a, b$ = semi‑major and semi‑minor axes                             |
|           |                                  | Hyperbola                          | [ $\frac{x^2}{a^2} - \frac{y^2}{b^2} = 1$ ]                                                                                | $a, b$ = real numbers defining hyperbola shape                      |
| 11        | **3D Geometry**                | Distance in 3D                     | [ $\sqrt{(x_2 - x_1)^2 + (y_2 - y_1)^2 + (z_2 - z_1)^2}$ ]                                                                 | Distance between points $(x_1,y_1,z_1)$ and $(x_2,y_2,z_2)$        |
|           |                                  | Section Formula                   | [ $\left( \frac{m x_2 + n x_1}{m + n}, \frac{m y_2 + n y_1}{m + n}, \frac{m z_2 + n z_1}{m + n} \right)$ ]                | Divides segment in ratio $m:n$                                      |
| 11        | **Limits & Derivatives**       | Limit Laws                        | [ $\lim_{x \to a} f(x) = L$ ]                                                                                               | Limit of function $f$ as $x$ approaches $a$                          |
|           |                                  | Standard Limits                   | [ $\lim_{x \to 0} \frac{\sin x}{x} = 1$, $\lim_{x \to 0} \frac{1 - \cos x}{x^2} = \frac12$ ]                            | Standard trigonometric limits                                        |
|           |                                  | Derivatives                        | [ $\frac{d}{dx} x^n = n x^{\,n-1}$, $\frac{d}{dx} \sin x = \cos x$ ]                                                      | Derivative formulas                                                   |
| 11        | **Mathematical Reasoning**     | Logical Connectives               | And (∧), Or (∨), Not (¬), If‑then (→), Iff (↔)                                                                            | Basic logical operators                                               |
|           |                                  | Contrapositive                     | [ $\neg q \rightarrow \neg p$ ]                                                                                             | Negation of $p$ and $q$                                              |
| 11        | **Statistics**                 | Mean                              | [ $\bar{x} = \frac{\sum f_i x_i}{\sum f_i}$ ]                                                                                 | $x_i$ values, $f_i$ frequencies                                      |
|           |                                  | Variance                          | [ $\sigma^2 = \frac{\sum (x_i - \bar{x})^2}{n}$ ]                                                                           | $n$ = total number of observations                                     |
|           |                                  | Standard Deviation                | [ $\sigma = \sqrt{\sigma^2}$ ]                                                                                               | Square root of variance                                                |
| 11        | **Probability**               | Basic Probability                 | [ $P(A) = \frac{\text{Favourable}}{\text{Total}}$ ]                                                                           | Probability of event $A$                                               |
|           |                                  | Complement                        | [ $P(A') = 1 - P(A)$ ]                                                                                                    | Probability of complement of $A$                                       |
|           |                                  | Addition Rule                     | [ $P(A \cup B) = P(A) + P(B) - P(A \cap B)$ ]                                                                               | Probability of union of $A$ and $B$                                      |
|           |                                  | Independence                     | [ $P(A \cap B) = P(A) \cdot P(B)$ ]                                                                                         | For independent events $A$ and $B$                                      |
| 12        | **Relations & Functions**      | Types of Relations                 | Reflexive: $(a,a)\in R$ <br> Symmetric: If $(a,b)\in R$ then $(b,a)\in R$ <br> Transitive: If $(a,b),(b,c)\in R$ then $(a,c)\in R$ | $a,b,c$ are elements in set; $R$ is a relation defined on the set        |
|           |                                  | Equivalence Relation              | Relation which is reflexive, symmetric, and transitive                                                                     | Same as above                                                             |
|           |                                  | Functions                          | One‑one, Onto, Inverse Function, Composite Function                                                                         | Functions mapping elements from one set to another                         |
| 12        | **Inverse Trigonometric Functions** | Principal Values             | [ $\sin^{-1}x$, $\cos^{-1}x$, $\tan^{-1}x$ defined on restricted domains ]                                                | $x$ is a real number within the domain of respective functions                |
|           |                                  | Properties                         | [ $\sin^{-1}x + \cos^{-1}x = \frac{\pi}{2}$, <br> $\tan^{-1}x + \cot^{-1}x = \frac{\pi}{2}$ ]                              | $x$ is a real number; $\pi$ is the constant pi                                  |
|           |                                  | Addition Formulas                  | [ $\tan^{-1}x + \tan^{-1}y = \tan^{-1}\!\left( \frac{x+y}{1-xy} \right)$ (if $xy<1$) ]                                  | $x$, $y$ are real numbers; domain condition: $xy < 1$                           |
| 12        | **Matrices**                    | Types of Matrices                  | Square, Diagonal, Scalar, Identity, Zero Matrix                                                                             | Classifications of matrix types                                                |
|           |                                  | Matrix Operations                  | Addition, Subtraction, Multiplication, Transpose                                                                            | Basic matrix operations                                                        |
|           |                                  | Determinant                        | For \(2\times2\): [ $|A| = ad - bc$ ]; For \(3\times3\): Expansion by minors                                               | $A$ is matrix; $a,b,c,d$ are entries of matrix                                |
|           |                                  | Inverse Matrix                    | [ $A^{-1} = \frac1{|A|}\mathrm{adj}\,A$ (if $|A|\neq0$) ]                                                                    | $A^{-1}$ is inverse of matrix $A$; $|A|$ determinant; adj = adjoint            |

|           |                                  | Expansion                          | Expansion along row or column                                                                                               | Method to compute determinant                                                |

| 12        | **Continuity & Differentiability** | Continuity                     | [ $f$ continuous at $x=a$ if $\lim_{x\to a}f(x) = f(a)$ ]                                                                | $f$ is function; $x\to a$ means approaching $a$                                  |
|           |                                  | Differential                      | $f$ differentiable at $x=a$ if derivative exists; differentiability ⇒ continuity                                          | $f'(x)$ derivative of $f(x)$                                                   |
|           |                                  | Derivative Formulas               | [ $(x^n)' = n x^{\,n-1}$, $(\sin x)' = \cos x$, $(\ln x)' = \tfrac1x$ ]                                                   | Standard derivatives; $x$ variable, $n$ constant                               |

|           |                                  | Incr/Decr                         | [ $f'(x) > 0$ increasing, $f'(x) < 0$ decreasing ]                                                                          | $f'(x)$ first derivative                                                       |
|           |                                  | Maxima & Minima                   | [ $f'(x) = 0$ and sign changes of $f'(x)$ for maxima/minima ]                                                              | $f'(x)$ is first derivative                                                   |
| 12        | **Integrals**                   | Indefinite Integral               | [ $\int x^n dx = \frac{x^{\,n+1}}{n+1} + C$, $\int e^x dx = e^x + C$ ]                                                   | $n\neq -1$, $C$ is constant of integration                                     |
|           |                                  | Definite Integral                 | [ $\int_a^b f(x)\,dx = F(b) - F(a)$ where $F'(x) = f(x)$ ]                                                                  | $F'(x) = f(x)$, $a,b$ are integration limits                                    |
|           |                                  | Properties                        | Linearity, additivity over intervals                                                                                         | Basic integral properties                                                      |

|           |                                  | Area between curves              | [ $\int_a^b (f(x) - g(x))\,dx$, where $f(x) \ge g(x)$ ]                                                                  | $f(x)$ upper curve; $g(x)$ lower curve                                        |
| 12        | **Differential Equations**       | Order and Degree                  | Order: highest derivative order; Degree: power of highest derivative                                                       | Definitions of order & degree                                                  |
|           |                                  | Separable Equations               | [ $\frac{dy}{dx} = g(x)\,h(y)$ ]                                                                                             | $y$ dependent variable; $x$ independent                                         |
