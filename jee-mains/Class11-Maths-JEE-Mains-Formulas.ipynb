| **Lesson** | **Concept** | **Formula / Key Point** |
|------------|-------------|--------------------------|
| **Sets** | Union | $A \cup B = \{ x : x \in A \text{ or } x \in B \}$ <br><br> |
|   | --- | --- |
|  | Intersection | $A \cap B = \{ x : x \in A \text{ and } x \in B \}$ <br><br> |
|   | --- | --- |
|  | Complement | $A' = U - A$ <br><br> |
|   | --- | --- |
|  | Cardinality | $n(A \cup B) = n(A) + n(B) - n(A \cap B)$ <br><br> |
|   | --- | --- |
| **Relations & Functions** | Function | A relation where every input has exactly one output <br><br> |
|   | --- | --- |
|  | Domain and Range | Set of inputs (domain), outputs (range) <br><br> |
|   | --- | --- |
|  | Types of Functions | One-one, onto, bijective <br><br> |
|   | --- | --- |
| **Trigonometry** | Identities | $\sin^2x + \cos^2x = 1$, $1 + \tan^2x = \sec^2x$, $1 + \cot^2x = \csc^2x$ <br><br> |
|   | --- | --- |
|  | Angle Formulas | $\sin(A \pm B) = \sin A \cos B \pm \cos A \sin B$ <br><br> |
|   | --- | --- |
|  | Double Angle | $\sin 2A = 2\sin A \cos A$, $\cos 2A = \cos^2A - \sin^2A$ <br><br> |
|   | --- | --- |
| **Complex Numbers** | General Form | $z = a + ib$ <br><br> |
|   | --- | --- |
|  | Conjugate | $\bar{z} = a - ib$ <br><br> |
|   | --- | --- |
|  | Modulus | $|z| = \sqrt{a^2 + b^2}$ <br><br> |
|   | --- | --- |
| **Quadratic Equations** | Roots | $x = \frac{-b \pm \sqrt{b^2 - 4ac}}{2a}$ <br><br> |
|   | --- | --- |
|  | Discriminant | $D = b^2 - 4ac$ <br><br> |
|   | --- | --- |
|  | Sum & Product of Roots | $\alpha + \beta = -\frac{b}{a}$, $\alpha \beta = \frac{c}{a}$ <br><br> |
|   | --- | --- |
| **Sequence & Series** | AP nth Term | $a_n = a + (n - 1)d$ <br><br> |
|   | --- | --- |
|  | AP Sum | $S_n = \frac{n}{2} [2a + (n - 1)d]$ <br><br> |
|   | --- | --- |
|  | GP nth Term | $a_n = ar^{n-1}$ <br><br> |
|   | --- | --- |
|  | GP Sum (finite) | $S_n = a \frac{r^n - 1}{r - 1}$, $r \ne 1$ <br><br> |
|   | --- | --- |
|  | GP Sum (infinite) | $S_{\infty} = \frac{a}{1 - r}$, $|r| < 1$ <br><br> |
|   | --- | --- |
| **Permutation & Combination** | Permutation | $_nP_r = \frac{n!}{(n - r)!}$ <br><br> |
|   | --- | --- |
|  | Combination | $_nC_r = \frac{n!}{r!(n - r)!}$ <br><br> |
|   | --- | --- |
|  | Property | $_nC_r = _nC_{n - r}$ <br><br> |
|   | --- | --- |
| **Binomial Theorem** | Expansion | $(x + a)^n = \sum_{k=0}^{n} \binom{n}{k} x^{n-k} a^k$ <br><br> |
|   | --- | --- |
|  | General Term | $T_{k+1} = \binom{n}{k} x^{n-k} a^k$ <br><br> |
|   | --- | --- |
| **Straight Lines** | Slope | $m = \frac{y_2 - y_1}{x_2 - x_1}$ <br><br> |
|   | --- | --- |
|  | Point-Slope Form | $y - y_1 = m(x - x_1)$ <br><br> |
|   | --- | --- |
|  | Slope-Intercept Form | $y = mx + c$ <br><br> |
|   | --- | --- |
|  | Distance | $\sqrt{(x_2 - x_1)^2 + (y_2 - y_1)^2}$ <br><br> |
|  | Angle Between Lines | $\tan \theta = \left| \frac{m_1 - m_2}{1 + m_1 m_2} \right|$ <br><br> |
|   | --- | --- |
| **Conic Sections** | Circle | $(x - h)^2 + (y - k)^2 = r^2$ <br><br> |
|   | --- | --- |
|  | Parabola | $y^2 = 4ax$, $x^2 = 4ay$ <br><br> |
|   | --- | --- |
|  | Ellipse | $\frac{x^2}{a^2} + \frac{y^2}{b^2} = 1$ <br><br> |
|   | --- | --- |
|  | Hyperbola | $\frac{x^2}{a^2} - \frac{y^2}{b^2} = 1$ <br><br> |
|   | --- | --- |
| **3D Geometry** | Distance in 3D | $\sqrt{(x_2 - x_1)^2 + (y_2 - y_1)^2 + (z_2 - z_1)^2}$ <br><br> |
|   | --- | --- |
|  | Section Formula | $\left( \frac{mx_2 + nx_1}{m+n}, \frac{my_2 + ny_1}{m+n}, \frac{mz_2 + nz_1}{m+n} \right)$ <br><br> |
|   | --- | --- |
| **Limits & Derivatives** | Limit Laws | $\lim_{x \to a} f(x) = L$ <br><br> |
|   | --- | --- |
|  | Standard Limits | $\lim_{x \to 0} \frac{\sin x}{x} = 1$, $\lim_{x \to 0} \frac{1 - \cos x}{x^2} = \frac{1}{2}$ <br><br> |
|   | --- | --- |
|  | Derivatives | $\frac{d}{dx} x^n = nx^{n-1}$, $\frac{d}{dx} \sin x = \cos x$ <br><br> |
|   | --- | --- |
| **Mathematical Reasoning** | Logical Connectives | And (∧), Or (∨), Not (¬), If-then (→), Iff (↔) <br><br> |
|   | --- | --- |
|  | Contrapositive | $\neg q \rightarrow \neg p$ <br><br> |
|   | --- | --- |
| **Statistics** | Mean | $\bar{x} = \frac{\sum f_ix_i}{\sum f_i}$ <br><br> |
|   | --- | --- |
|  | Variance | $\sigma^2 = \frac{\sum (x_i - \bar{x})^2}{n}$ <br><br> |
|   | --- | --- |
|  | Standard Deviation | $\sigma = \sqrt{\sigma^2}$ <br><br> |
|   | --- | --- |
| **Probability** | Basic Probability | $P(A) = \frac{\text{Favorable}}{\text{Total}}$ <br><br> |
|   | --- | --- |
|  | Complement | $P(A') = 1 - P(A)$ <br><br> |
|   | --- | --- |
|  | Addition Rule | $P(A \cup B) = P(A) + P(B) - P(A \cap B)$ <br><br> |
|   | --- | --- |
|  | Independence | $P(A \cap B) = P(A) \cdot P(B)$ <br><br> |
